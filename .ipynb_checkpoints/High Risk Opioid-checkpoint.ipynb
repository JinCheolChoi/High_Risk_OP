```python
#***************
#
# Import package
#
#***************
# import packages
import pandas as pd
import matplotlib.pyplot as plt # plot
import numpy as np

import tensorflow as tf         #
from tensorflow import keras    #

import nltk # natural language processing toolkit

from wordcloud import WordCloud

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer

from nltk import word_tokenize          
from nltk.stem import WordNetLemmatizer 

from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#**************************
#
# Define required functions
#
#**************************
def lemma(data, column):
    lemmatized_sentence=[]
    
    #temp_data=data[column].reset_index()
    #for i in data[column].index:
    for i in range(data.shape[0]):
        wnl = WordNetLemmatizer()
        sentence=data[column][i]
        lemma_sentence = ' '.join([wnl.lemmatize(wnl.lemmatize(t,'v'),'n') for t in word_tokenize(sentence)])
        lemmatized_sentence.append(lemma_sentence)   
    data[column+'_Lemmtzd']=lemmatized_sentence  
    
def bag_of_words_model(data, column):
    cv = CountVectorizer()
    bag_of_words=cv.fit_transform(data[column])
    feature_names = cv.get_feature_names()

    return(pd.DataFrame(bag_of_words.toarray(), columns = feature_names))

def tfidf(dataframe):
    # 1st way (using vector of full sentence)
    #tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 2), stop_words='english')
    #features = tfidf.fit_transform(data[column]).toarray()
    
    # 2nd way (using 'bag of words model')
    tfidf_transformer = TfidfTransformer()
    features = tfidf_transformer.fit_transform(dataframe)
    feature_names = list(dataframe)
    
    return(pd.DataFrame(features.toarray(), columns = feature_names))

def print_class_prediction(sentence, clf):
    '''For a sentence and classifier, classify the sentence and then show the estimated probabilities
    for each class as a bar chart.
    '''
    # lemmatize sentence
    wnl = WordNetLemmatizer()
    lam_sentence=' '.join([wnl.lemmatize(wnl.lemmatize(t,'v'),'n') for t in word_tokenize(sentence)])
    # generate 'bag of words' model for it
    cv = CountVectorizer()
    bag_of_words=cv.fit_transform(pd.Series(lam_sentence))
    feature_names = cv.get_feature_names()
    # list of all column names in the entire data
    lists=list(X_train)
    temp_DF=pd.DataFrame(np.zeros(shape=(1, len(lists))), columns=lists)
    
    #
    in_columns = []
    for i in range(0, len(lists)):
        in_columns.append(lists[i] in feature_names)
    
    temp_DF.iloc[:, np.where(in_columns)[0]]=pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
    
    #
    if clf==text_nn_clf:
        probs = np.array([1-text_nn_clf.predict(temp_DF).flatten(), text_nn_clf.predict(temp_DF).flatten()]).flatten()
    else :
        probs = clf.predict_proba(temp_DF).flatten()
    
    classes = le.classes_
    x = np.arange(len(classes))

    plt.figure(figsize=(10,6));
    plt.grid(True,axis='y');
    plt.bar(x,probs);
    plt.xticks(x,classes,rotation=45,ha='right',fontsize=14);

    ind = np.argmax(probs)
    max_prob = probs[ind]
    max_class = classes[ind]
    plt.title('Predicted condition : {}, confidence {:.2%}'.format(max_class, max_prob), fontsize=14);

#def print_class_prediction(sentence, clf):
#    '''For a sentence and classifier, classify the sentence and then show the estimated probabilities
#    for each class as a bar chart.
#    '''
#    probs = clf.predict_proba([sentence]).flatten()
#    
#    classes = le.classes_
#    x = np.arange(len(classes))
#
#    plt.figure(figsize=(10,6));
#    plt.grid(True,axis='y');
#    plt.bar(x,probs);
#    plt.xticks(x,classes,rotation=45,ha='right',fontsize=14);
#
#    ind = np.argmax(probs)
#    max_prob = probs[ind]
#    max_class = classes[ind]
#    plt.title('Predicted condition : {}, confidence {:.2%}'.format(max_class, max_prob), fontsize=14);

def create_word_cloud_from_HIGH_FQ(df, HIGH_FQ=1):
    '''Create a word cloud from HIGH_FQ
    
    '''
    text = df[df['HIGH_FQ'] == HIGH_FQ]['ORDER_DIRECTIONS'].values
    text = ' '.join(text)
    word_cloud = WordCloud(width=600,height=400,
                       max_words=150, background_color="white",
                       collocations=False).generate(text)
    plt.figure(figsize=(12,10));
    plt.imshow(word_cloud, interpolation='bilinear');
    plt.axis('off');


class print_resampled_accuracy():
    def __init__(self, predicted, y, samples=1000):
        '''Bootstrap confidence intervals for test accuracy
        '''
        from sklearn.utils import resample
        acc = []
        for _ in range(samples):
            r_p, r_y = resample(predicted, y)
            acc.append(np.mean(r_p == r_y))
            
            self.lc ,self.m, self.uc = np.percentile(acc,[2.5,50.,97.5]);
```

```python
#************
#
# Import data
#
#************
# designate directory path
data_dir="/Users/JinCheol/Desktop/BCCSU/PROJECTS/Mint's project/Natural Language Process/Data/"

# import files
BL_Q_Orgnl=pd.read_csv(data_dir + "pharmacy_dat (only Opioid).csv")

#data=BL_Q_Orgnl.sample(50)
data=BL_Q_Orgnl

#***********
# word cloud
#***********
# only data with the outcome of 0
create_word_cloud_from_HIGH_FQ(data, 0)
# only data with the outcome of 1
create_word_cloud_from_HIGH_FQ(data, 1)
```

```python
#****************
#
# Data processing
#
#****************
# predictor variables
#********************
# generate lemmatized variables
lemma(data, 'ITEM_DESCRIPTION')
lemma(data, 'ORDER_DIRECTIONS')

## see the column 'ITEM_DESCRIPTION_Lemmtzd'
#data.iloc[:, np.where(np.array(list(data))=='ITEM_DESCRIPTION_Lemmtzd')[0][0]]
#data.iloc[:, np.where(np.array(list(data))=='ITEM_DESCRIPTION')[0][0]]

# generate 'bag of words model' with lemmatized variables
bag_of_words_item_description=bag_of_words_model(data, 'ITEM_DESCRIPTION_Lemmtzd')
bag_of_words_order_directions=bag_of_words_model(data, 'ORDER_DIRECTIONS_Lemmtzd')

# obtain tfidf weights on each 'bag of words' models
tfidf_weights_item_description=tfidf(bag_of_words_item_description)
tfidf_weights_order_directions=tfidf(bag_of_words_order_directions)

#
tfidf_weights_item_description=bag_of_words_item_description
tfidf_weights_order_directions=bag_of_words_order_directions

# combine all data of tfidf weights
tfidf_weights=pd.concat([tfidf_weights_item_description, tfidf_weights_order_directions], axis=1, ignore_index=True)
tfidf_weights=pd.concat([tfidf_weights_item_description, tfidf_weights_order_directions], axis=1)

# remove stop words
tokenized_words=list(tfidf_weights)
stop_words=stopwords.words('english')
info_words=[word for word in tokenized_words if word not in stop_words]
tfidf_weights=tfidf_weights[info_words]
tfidf_weights=tfidf_weights.loc[:,~tfidf_weights.columns.duplicated()]
```

```python
#**********************************************************
# response variable (into which data are to be categorized)
#**********************************************************
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
target=le.fit_transform(data['HIGH_FQ'].values)
```

```python
#**************************
#
# Create test / train split
#
#**************************
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    #bag_of_words
    tfidf_weights, 
    target, test_size=0.10, random_state=42)
print('Training data size: {}'.format(len(X_train)))
print('Testing data size: {}'.format(len(X_test)))
```

```python
#***********************
#
# Naive Bayes Classifier
#
#***********************
from sklearn.naive_bayes import MultinomialNB, GaussianNB
# Multinomial NB
multi_NB_clf = MultinomialNB()
multi_NB_clf.fit(X_train, y_train)

# guassian NB
gauss_NB_clf = GaussianNB()
gauss_NB_clf.fit(X_train, y_train)


# predict on the validation data and calculate the accuracy
# multinomial NB
predicted = multi_NB_clf.predict(X_train)
tr_acc=print_resampled_accuracy(predicted, y_train)
print('[multi_NB] Training accuracy : {:.2%} ({:.2%} - {:.2%})'.format(tr_acc.m, tr_acc.lc, tr_acc.uc))
predicted = multi_NB_clf.predict(X_test)
tst_acc=print_resampled_accuracy(predicted, y_test)
print('[multi_NB] Test accuracy : {:.2%} ({:.2%} - {:.2%})'.format(tst_acc.m, tst_acc.lc, tst_acc.uc))

# gaussian NB
predicted = gauss_NB_clf.predict(X_train)
tr_acc=print_resampled_accuracy(predicted, y_train)
print('[gauss_NB] Training accuracy : {:.2%} ({:.2%} - {:.2%})'.format(tr_acc.m, tr_acc.lc, tr_acc.uc))
predicted = gauss_NB_clf.predict(X_test)
tst_acc=print_resampled_accuracy(predicted, y_test)
print('[gauss_NB] Test accuracy : {:.2%} ({:.2%} - {:.2%})'.format(tst_acc.m, tst_acc.lc, tst_acc.uc))
```

```python
#**************
#
# Random Forest
#
#**************
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=100, random_state=0)
rf_clf.fit(X_train, y_train)

# predict on the validation data and calculate the accuracy
# multinomial NB
predicted = rf_clf.predict(X_train)
tr_acc=print_resampled_accuracy(predicted, y_train)
print('[RF] Training accuracy : {:.2%} ({:.2%} - {:.2%})'.format(tr_acc.m, tr_acc.lc, tr_acc.uc))
predicted = rf_clf.predict(X_test)
tst_acc=print_resampled_accuracy(predicted, y_test)
print('[RF] Test accuracy : {:.2%} ({:.2%} - {:.2%})'.format(tst_acc.m, tst_acc.lc, tst_acc.uc))
```

```python
#***************
#
# Neural Network
#
#****************
# Build the model
#****************
#text_nn_clf = keras.Sequential([
#    keras.layers.Dense(128, activation=tf.nn.relu), # the first dense layer has 128 nodes
#    keras.layers.Dense(10, activation=tf.nn.relu), # the second layer of 10 nodes with the activation function of softmax
#    keras.layers.Dense(1, activation=tf.nn.sigmoid)
#])
text_nn_clf = keras.Sequential()

text_nn_clf.add(keras.layers.Dense(4, activation=tf.nn.relu)) # dense layer with 16 hidden units
text_nn_clf.add(keras.layers.Dense(4, activation=tf.nn.relu)) # dense layer with 16 hidden units
text_nn_clf.add(keras.layers.Dense(4, activation=tf.nn.relu)) # dense layer with 16 hidden units
text_nn_clf.add(keras.layers.Dense(1, activation=tf.nn.sigmoid)) 

text_nn_clf.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['acc'])
#****************
# Train the model
#****************
history = text_nn_clf.fit(np.array(X_train, dtype=np.float64),
                    np.array(y_train, dtype=np.float64),
                    epochs=30,
                    validation_data=(np.array(X_test, dtype=np.float64), np.array(y_test, dtype=np.float64)),
                    verbose=2)
#*******************
# Evaluate the model
#*******************
results = text_nn_clf.evaluate(X_test, y_test)
print(results)
```

```python
#**************************************************
# Generate a graph of accuracy and loss over epochs
#**************************************************
history_dict = history.history
print("There are four entries as {}".format(history_dict.keys()))

# import pyplot
import matplotlib.pyplot as plt

acc = history_dict['acc']
val_acc = history_dict['val_acc']
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(acc) + 1)

# "bo" is for "blue dot"
plt.plot(epochs, loss, 'bo', label='Training loss')
# b is for "solid blue line"
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
